## What is mask() in Pandas? ðŸ§ 

### mask() conditionally replaces values in a Series or DataFrame.

#### If the condition is True, the value is replaced.
#### If the condition is False, the value stays as it is.

Thatâ€™s the core idea.

Basic syntax
df.mask(condition, new_value)


condition â†’ boolean (True / False) for each cell

new_value â†’ what to replace with (can be a value, NaN, or another Series)

Simple example (numbers)
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50])

Mask values greater than 30
s.mask(s > 30, 999)

Output
0     10
1     20
2     30
3    999
4    999


âœ”ï¸ Values where condition is True are replaced
